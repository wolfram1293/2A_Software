```python
def rand_val(minval=0.0,maxval=1.0):
    x = 1
    a = 1664525
    c = 1013904223
    m = 4294967296
    while True:
        xnext = (a * x + c) % m
        yield minval + (maxval - minval) * xnext / m
        x = xnext
```

```python
r = rand_val(minval=-5.0,maxval=5.0)
for i in range(10):
    print(next(r))
```

```python
def logistic(a=4,x0=1):
    if a<0 or a>4 or x0<0 or x0>1:
        raise ValueError('Error')
    else:
        x=x0
        while True:
            xnext=a*x*(1-x)
            yield xnext
            x = xnext
```

```python
g = logistic(a=4,x0=1)
print(next(g))
```

```python
def mean(l):
    if len(l) == 0:
        raise ValueError('Length of list is 0')
    else:
        s = 0.0
        size = len(l)
        for i in range(size):
            s += l[i]
        return s / size
l = [1,2,3]
```

```python
mean(l)
```

```python
import random
def random_walk(x0=0,p=0.5):
    if p<0 or p>1:
        raise ValueError('Error')
    else:
        x=x0
        while True:
            r=random.random()
            if r<p:
                x=x+1
            else:
                x=x-1
            yield x
```

```python
import matplotlib.pyplot as plt
r=random_walk(p=0.51)
N=1000
for i in range(N):
        y=next(r)
        plt.scatter(i,y,s=0.8,color='b')
```

```python
import sympy

def taylor(f,x,a,n):
    A=f.subs(x,a)
    n1=1
    dfdx=f
    for i in range(n):
        if i!=0:
            n1=n1*i
            dfdx=sympy.diff(dfdx,x)
            B=dfdx.subs(x,a)*(x-a)**i/n1
            A+=B
    return A
```

```python
sympy.init_printing()
x = sympy.Symbol('x')
taylor(sympy.sin(x),x,0,8)
```
